# Q8 Split the pima_df into training and test set in the ratio of 70:30 (Training:Test).</b>


```python
from sklearn.model_selection import train_test_split

array = df.values
# We know the data that's why taking 8 columns. alternate way is to assign full dataframe and drop class column. Either way will work
X = array[:,0:7]
Y = array[:,8]  
test_size = 0.30 
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=1)
```

# Q9 Create the decision tree model using â€œentropyâ€ method of reducing the entropy and fit it to training data.</b>



```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import metrics

dt_model = DecisionTreeClassifier(criterion = 'entropy' )

print(dt_model.fit(X_train, y_train))

# As per requirement of the question, data has been fit in the model. Now we are trying to plot the decision tree

######## This is code for plotting the decision tree in this given scenario
##########################################################################
from sklearn.externals.six import StringIO
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dt_model, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

# Q10 Print the accuracy of the model & print the confusion matrix 

```python
print(dt_model.score(X_test , y_test))
y_predict = dt_model.predict(X_test)
print(metrics.confusion_matrix(y_test, y_predict))
```
